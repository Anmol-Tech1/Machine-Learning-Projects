### Import necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Data collection and Data preprocessing

```python
sonar_data = pd.read_csv("sonar data.csv", header=None)
sonar_data.head()
```

```python
# Number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
# Saperating data and labels
X = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]
```

### Training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify=y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Model training

```python
model = LogisticRegression()
```

```python
# traininf the logistic regression model with the training data
model.fit(X_train, y_train)
```

### Model Evaluation

```python
# Accurary on training data
X_train_prediction = model.predict(X_train)
training_data_accurary = accuracy_score(X_train_prediction, y_train)
```

```python
print("Accuracy on training data:", training_data_accurary)
```

```python
# Accurary score of test data
X_test_prediction = model.predict(X_test)
testing_data_accurary = accuracy_score(X_test_prediction, y_test)
```

```python
print("Accuracy on test data: ", testing_data_accurary)
```
